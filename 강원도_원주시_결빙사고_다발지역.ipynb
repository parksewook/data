```python
import pandas as pd
import json
import requests
from pandas.io.json import json_normalize

addr = ['2017', '2018', '2019', '2020']
j = 0

for i in addr :
  url = ("http://apis.data.go.kr/B552061/frequentzoneFreezing/getRestFrequentzoneFreezing?ServiceKey=sgqqn0iw%2BOkWOrg1ZV5X6mVHgaFoGbXsbLEwh2p3bKPZhWLXKDPdR%2F49%2FtVt%2F3XQbRur9ZiwPomz4Md%2B%2BZ0FVg%3D%3D&siDo=42&guGun=130&numOfRows=100&pageNo=1&type=json&searchYearCd=" + i)
  response = requests.get(url)
  contents = response.text
  json_ob = json.loads(contents)
  body = json_ob['items']['item']
  dataframe = json_normalize(body)
  addr[j] = dataframe
  j = j + 1

acid_2017 = addr[0][["sido_sgg_nm", "spot_nm", "occrrnc_cnt", "caslt_cnt", "dth_dnv_cnt", "se_dnv_cnt", "sl_dnv_cnt", "wnd_dnv_cnt", "lo_crd", "la_crd" ]]
acid_2018 = addr[1][["sido_sgg_nm", "spot_nm", "occrrnc_cnt", "caslt_cnt", "dth_dnv_cnt", "se_dnv_cnt", "sl_dnv_cnt", "wnd_dnv_cnt", "lo_crd", "la_crd" ]]
acid_2019 = addr[2][["sido_sgg_nm", "spot_nm", "occrrnc_cnt", "caslt_cnt", "dth_dnv_cnt", "se_dnv_cnt", "sl_dnv_cnt", "wnd_dnv_cnt", "lo_crd", "la_crd" ]]
acid_2020 = addr[3][["sido_sgg_nm", "spot_nm", "occrrnc_cnt", "caslt_cnt", "dth_dnv_cnt", "se_dnv_cnt", "sl_dnv_cnt", "wnd_dnv_cnt", "lo_crd", "la_crd" ]]

acid_2017.rename(columns = {'sido_sgg_nm':'시도시군구명', 'spot_nm': '지점명', 'occrrnc_cnt':'발생건수', 'caslt_cnt':'사상자수', 'dth_dnv_cnt':'사망자수', 'se_dnv_cnt':'중상자수', 'sl_dnv_cnt':'경상자수', 'wnd_dnv_cnt':'부상신고자수'}, inplace=True)
acid_2018.rename(columns = {'sido_sgg_nm':'시도시군구명', 'spot_nm': '지점명', 'occrrnc_cnt':'발생건수', 'caslt_cnt':'사상자수', 'dth_dnv_cnt':'사망자수', 'se_dnv_cnt':'중상자수', 'sl_dnv_cnt':'경상자수', 'wnd_dnv_cnt':'부상신고자수'}, inplace=True)
acid_2019.rename(columns = {'sido_sgg_nm':'시도시군구명', 'spot_nm': '지점명', 'occrrnc_cnt':'발생건수', 'caslt_cnt':'사상자수', 'dth_dnv_cnt':'사망자수', 'se_dnv_cnt':'중상자수', 'sl_dnv_cnt':'경상자수', 'wnd_dnv_cnt':'부상신고자수'}, inplace=True)
acid_2020.rename(columns = {'sido_sgg_nm':'시도시군구명', 'spot_nm': '지점명', 'occrrnc_cnt':'발생건수', 'caslt_cnt':'사상자수', 'dth_dnv_cnt':'사망자수', 'se_dnv_cnt':'중상자수', 'sl_dnv_cnt':'경상자수', 'wnd_dnv_cnt':'부상신고자수'}, inplace=True)

acid_2017["발생년도"] = len(acid_2017) * ["2013 ~ 2017"]
acid_2018["발생년도"] = len(acid_2018) * ["2014 ~ 2018"]
acid_2019["발생년도"] = len(acid_2019) * ["2015 ~ 2019"]
acid_2020["발생년도"] = len(acid_2020) * ["2016 ~ 2020"]
```
